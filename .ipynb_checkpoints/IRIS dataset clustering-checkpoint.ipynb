```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
# Importing the dataset 
data = pd.read_csv("D:\sem-6\DSPL\DSPL_LAB-main\Data_Set\IRIS.csv")
data
```

```python
data["species"].replace({"Iris-setosa":"0", "Iris-versicolor":"1" , "Iris-virginica": "2"}, inplace = True) 
data
```

```python
# Using K-Means Clustering
x = data.iloc[:,2:4] # 1st for rows and second for columns
x
```

```python
 #  this fir(x) => taking the 2 input as col  x and col y as an input value and then it will predict the slutering by calculating the means 
kmeans = KMeans(3) # Creating three clusters
kmeans.fit(x) 
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
# printing the predicted cluster
```

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['petal_length'],data_with_clusters['petal_width'],c=data_with_clusters['Clusters'],cmap='rainbow')
# here c takes the target value or it is also called the class label
```
